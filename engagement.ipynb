```python
import pandas as pd
import numpy as np
```

```python
df_geo = pd.read_csv(r"NatGeo Scrape 11-27.csv")
df_geo
```

```python
max_likes=df_geo.likeCount.max()
max_comment=df_geo.commentCount.max()
df_geo['normalizedCommentCount']=df_geo['commentCount']/max_comment #divide by number of comments
df_geo['normalizedLikeCount']=df_geo['likeCount']/max_likes #divide by number of max likes
df_geo['engagementScore']=0.4*df_geo['normalizedLikeCount'] + 0.6*df_geo['normalizedCommentCount']
medianEngagementScore= df_geo['engagementScore'].median() # median number for engagement
df_geo['engagement']= df_geo['engagementScore'].apply(lambda x: 1 if x > medianEngagementScore else 0) #if higher than median then 1
df_geo
```
